```python
# This means we only have valid 7437 cf_ratings and 10280 cc_ratings,let us find out where we have both the ratings
df_1=df.dropna()
print(df_1.shape)
y=df_1['cf_rating']
x=df_1['cc_rating']
```

```python
plt.grid()
plt.scatter(x,y,color='blue')
plt.xlabel('CodeChef Rating')
plt.ylabel('CodeForces Rating')
plt.show()
```

```python
sns.regplot(x=x,y=y,scatter_kws={'color':'blue'},line_kws={'color':'red'})
plt.grid()
plt.xlabel('CodeChef Rating')
plt.ylabel('CodeForces Rating')
plt.title('Regression Plot')
plt.show()
```

```python
def linear_regression(X, y):
    n = len(X)
    mean_x = sum(X) / n
    mean_y = sum(y) / n
    numer = sum((X.iloc[i] - mean_x) * (y.iloc[i] - mean_y) for i in range(n))
    denom = sum((X.iloc[i] - mean_x) ** 2 for i in range(n))
    slope = numer / denom if denom != 0 else 0
    intercept = mean_y - slope * mean_x
    return slope, intercept

m,b=linear_regression(x,y)
```

```python
print(m,b)
```

```python
def fun(codechef_rating):
    codeforces_rating=(codechef_rating*m)+b
    return codeforces_rating

for z in range(700,2500,50):
    print("Codechef :",z,"Codeforces :",round(fun(z),2))
```
